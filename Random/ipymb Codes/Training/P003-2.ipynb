## find transpose 
#### d1=[[12,6,1],[5,9,2],[7,15,3],[13,2,4]]

```python
d=[[12,5,7,13],[6,9,15,2],[1,2,3,4]]
```

```python
d1=[x for x in range(3)]
d1
```

```python
d2=[y for y in range(4)]
d2
```

```python
d3=[[d[x][y] for x in range(len(d))] for y in range(len(d[0]))]
d3
```

```python
d=[[12,5,7,13],[6,9,15,2],[1,2,3,4]]
d
```

```python
d1=[[d[i][j] for i in range(len(d))] for j in range(len(d[0])) ]
d1
```
